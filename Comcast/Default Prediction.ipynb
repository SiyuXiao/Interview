```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelBinarizer
from sklearn.neural_network import MLPClassifier
train = pd.read_csv('train data and label.csv')
train = train.fillna('')
train_data = train.loc[:, 'industry':'insured_amount']
test = pd.read_csv('test data.csv')
test = test.fillna('')
test_data = test.loc[:, 'industry':'insured_amount']
data = pd.concat([train_data , test_data])
data_onehot = OneHotEncoder().fit_transform(data).toarray()
data_onehot.shape
```

```python
train_data_onehot = data_onehot[:3006]
train_data_onehot.shape
```

```python
test_data_onehot = data_onehot[3006:]
test_data_onehot.shape
```

```python
train_label = train[['default_status']]
train_label_onehot = LabelBinarizer().fit_transform(train_label)
```

```python
dtc = tree.DecisionTreeClassifier(criterion="entropy")
clf = dtc.fit(train_data_onehot, train_label_onehot)
```

```python
predict = clf.predict(test_data_onehot)
predict = predict.reshape(1, -1)
```

```python
test_label = pd.read_csv('test label.csv', nrows=1288)
test_label = test_label[['default_status']]
test_label = test_label.values.reshape(1, -1)
test_label.shape
```

```python
'Accuracy: {0}'.format((predict == test_label).sum() / len(predict[0]))
```

```python
mlp = MLPClassifier(alpha=1e-05, hidden_layer_sizes=(128, 32), random_state=1, solver='lbfgs')
clf = mlp.fit(train_data_onehot, train_label_onehot.reshape(-1))
```
